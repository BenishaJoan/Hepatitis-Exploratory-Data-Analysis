### Import the necessary packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import the dataset

```python
df = pd.read_csv('C:/Users/benis/Desktop/Benisha_Joan/Dataset/hepatitis_csv.csv')
df.head(10)
```

### Exploratory Analysis

```python
df.shape
```

```python
df.info()
```

```python
### converting categorical values to numerical values
```

```python
replace_values = {'live':1,'die':0,False:0,True:1,
                 'male':0,'female':1}
```

```python
df.replace(replace_values,inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
### Finding the total no of patients
```

```python
total_patients = df.shape[0]
patients_live  = np.sum(df['class']==1)/total_patients * 100
patients_died  = np.sum(df['class']==0)/total_patients * 100

print("Percent of patients alive : ",round(patients_live,2))
print("Percent of patients died  : ",round(patients_died,2))
```

```python
sns.countplot(df['sex'])
print(df['sex'].value_counts())
```

```python
### dropping the null values
```

```python
df = df.dropna()
```

```python
f ,axes = plt.subplots(figsize=(15,6),nrows=1,ncols=2)
female = df[df['sex']==0]
ax = sns.distplot(female[female['class']==0].age,ax=axes[0],label='Died',hist=False)
ax = sns.distplot(female[female['class']==1].age,ax=axes[0],label='Alive',hist=False)
ax.legend()
ax.set_title('Female')

male = df[df['sex']==1]
ax1 = sns.distplot(male[male['class']==0].age,ax=axes[1],label='Died',hist=False)
ax1 = sns.distplot(male[male['class']==1].age,ax=axes[1],label='Alive',hist=False)
ax1.legend()
ax1.set_title('Male')

```

```python
f,ax = plt.subplots(figsize=(8,8))
sns.heatmap(df.corr(),linewidths=.3,annot =False,ax=ax)
```
